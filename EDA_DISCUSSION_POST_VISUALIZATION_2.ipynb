```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data into a DataFrame
data = pd.DataFrame({
    "Date": ["2001-08-08", "2001-08-09", "2001-08-10", "2001-08-13", "2001-08-14", "2001-08-15"],
    "Open": [28.184, 26.665, 26.571, 26.333, 25.906, 26.333],
    "High": [28.232, 26.997, 26.903, 26.427, 26.665, 26.949],
    "Low": [26.618, 26.143, 26.029, 25.479, 25.812, 26.286],
    "Close": [27.225, 26.911, 26.571, 25.906, 26.286, 26.647],
    "Volume": [5558323, 3158167, 3365553, 5068619, 3865059, 2763197],
    "OpenInt": [0, 0, 0, 0, 0, 0]
})

# Convert Date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Plot Closing Price over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Date'], data['Close'], marker='o', label='Closing Price')
plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Scatter Plot: Volume vs. Close
plt.figure(figsize=(8, 4))
plt.scatter(data['Volume'], data['Close'], alpha=0.7, label='Volume vs. Close')
plt.title("Volume vs. Closing Price")
plt.xlabel("Volume")
plt.ylabel("Closing Price")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
